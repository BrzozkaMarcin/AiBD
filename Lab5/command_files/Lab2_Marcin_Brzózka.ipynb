```python
# Zaimportowanie wymaganych bibliotek.
import numpy as np
import pandas as pd
import os
```

**Wczytanie pliku csv z danymi**

```python
data = pd.read_csv(r'../original_data/billboard.csv', encoding= 'unicode_escape')
data
```

**Stworzenie tabeli w nowej formie na dane**  
Pominięcie danej z rokiem

```python
better_data = pd.DataFrame({
    'artist': [data.values[0][1]], 'track': [data.values[0][2]],
    'time': [data.values[0][3]], 'genre': [data.values[0][4]],
    'date entered': [data.values[0][5]], 'date peaked': [data.values[0][6]],
    'week': int(1), 'rank': int(data.values[0][7])
})
```

**Uzupełnienie nowej tabeli**

```python
X, Y = data.shape
for i in range(X):
    for j in range(7, Y):
        if not np.isnan(data.iloc[i][j]) and (i != 0 or j != 7):
            X_1 = better_data.shape[0]
            better_data.loc[X_1, :] = [data.values[i][1], data.values[i][2], data.values[i][3], data.values[i][4],
                data.values[i][5], data.values[i][6], int(j-6), data.iloc[i][j]]
better_data
```

**Eksport danych do pliku z rozszerzeniem csv oraz xlsx**

```python
better_data.to_csv(r"../analysis_data/billboard_v2.csv", encoding= 'unicode_escape')
better_data.to_excel(r"../analysis_data/billboard_v2.xlsx", encoding= 'unicode_escape', index=False)
```
